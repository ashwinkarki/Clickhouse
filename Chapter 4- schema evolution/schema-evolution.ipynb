What is Schema Evolution?

Schema evolution means:

You can change the table structure (add/remove columns, modify types in some cases) without dropping the table or rewriting all existing data.

In ClickHouse, this is handled at column level, thanks to how MergeTree stores parts.



2️⃣ Basic Rules

You can add new columns easily

You can remove columns (only if no older parts reference them)

You can’t change column order — ClickHouse stores columns physically by name

You can’t easily change column type (requires ALTER + cast or create new column)

Default values are important for backward compatibility

```python
ALTER TABLE events
ADD COLUMN device_type String DEFAULT 'unknown';
```

```python
ALTER TABLE events
DROP COLUMN device_type;
```

```python
ALTER TABLE events
RENAME COLUMN event_type TO action_type;
```
