select * from users_db.users order by user_id limit 1000;

✅ system.metrics

Real-time server metrics

Memory usage

Number of threads

Query execution stats

```python
SELECT
    metric,
    value
FROM system.metrics
WHERE metric IN (
    'Query',
    'MemoryTracking',
    'GlobalThreadActive',
    'PartsActive',
    'Merge',
    'TCPConnection'
);
```

system.events

SELECT
    event,
    value
FROM system.events
WHERE event IN (
    'Query',
    'SelectQuery',
    'QueryTimeMicroseconds',
    'RealTimeMicroseconds',
    'SelectedRows',
    'RowsReadByMainReader',
    'SelectQueriesWithPrimaryKeyUsage',
    'DiskReadElapsedMicroseconds',
    'SynchronousRemoteReadWaitMicroseconds',
    'S3ReadRequestsErrors',
    'GlobalThreadPoolExpansions',
    'LocalThreadPoolJobWaitTimeMicroseconds',
    'Merge',
    'MarkCacheHits',
    'MarkCacheMisses'
)
ORDER BY event;

system.processes

this is a real-time running query monitor in ClickHouse.

system.processes shows currently running queries only.

⚠️ Once a query finishes → it disappears from this table.

SELECT
    query_id,
    user,
    elapsed,
    read_rows,
    formatReadableSize(read_bytes) AS read_size,
    formatReadableSize(memory_usage) AS memory_used,
    query
FROM system.processes
WHERE query LIKE '%users_db.users%'
ORDER BY elapsed DESC;
